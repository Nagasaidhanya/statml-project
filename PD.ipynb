```python
X=data.drop('price',axis=1)
y=data['price']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
data.head()
```

```python
#as we can see our target variable(price) has continuos values
#so it is regression.....
```

```python
#importing models....

from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
```

```python
#model training....
lr=LinearRegression()
lr.fit(X_train,y_train)

svm=SVR()
svm.fit(X_train,y_train)

rf=RandomForestRegressor()
rf.fit(X_train,y_train)

gbr=GradientBoostingRegressor()
gbr.fit(X_train,y_train)


xg=XGBRegressor()
xg.fit(X_train,y_train)
```

```python
#prediction of the data....

y_pred1=lr.predict(X_test)
y_pred2=svm.predict(X_test)
y_pred3=rf.predict(X_test)
y_pred4=gbr.predict(X_test)
y_pred5=xg.predict(X_test)
```

```python
#evaluating algo....

from sklearn import metrics
score1=metrics.r2_score(y_test,y_pred1)
score2=metrics.r2_score(y_test,y_pred2)
score3=metrics.r2_score(y_test,y_pred3)
score4=metrics.r2_score(y_test,y_pred4)
score5=metrics.r2_score(y_test,y_pred5)

```

```python
import numpy as np
import pandas as pd
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error
# Number of bootstrap iterations
num_iterations = 100

# Lists to store MSE values
mse_values = []

for i in range(num_iterations):
    # Create a random training and test set using bootstrap resampling
    indices = np.random.choice(len(X), size=len(X), replace=True)
    X_train = X.iloc[indices]
    y_train = y.iloc[indices]

    # Train an SVR model
    svm = SVR()
    svm.fit(X_train, y_train)

    # Make predictions
    y_pred = svm.predict(X_test)

    # Calculate Mean Squared Error (MSE)
    mse = mean_squared_error(y_test, y_pred)
    mse_values.append(mse)

# Calculate and print the mean and standard deviation of MSE
mean_mse = np.mean(mse_values)
std_mse = np.std(mse_values)
print(f"Mean of MSE across {num_iterations} iterations: {mean_mse}")
print(f"Standard Deviation of MSE: {std_mse}")
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
# Number of bootstrap iterations
num_iterations = 100
# Lists to store MSE and coefficients from each iteration
mse_values = []
coefficients_list = []
for i in range(num_iterations):
    # Create a random training set using bootstrap resampling
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=None, shuffle=True)
    alpha = 0.01
    lasso = Lasso(alpha=alpha)
    lasso.fit(X_train, y_train)
    y_pred = lasso.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    coefficients_list.append(lasso.coef_)
mean_mse = np.mean(mse_values)
std_mse = np.std(mse_values)
print(f"Mean of MSE across {num_iterations} iterations: {mean_mse}")
print(f"Standard Deviation of MSE: {std_mse}")
average_coefficients = np.mean(coefficients_list, axis=0)
print("Average Lasso Coefficients:", average_coefficients)
```
